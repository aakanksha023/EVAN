```python
df_x = df.groupby(["FOLDERYEAR","BusinessIndustry",'BusinessType'],as_index=False)[["label"]].count()
df_x.columns = ['FOLDERYEAR', 'BusinessIndustry','BusinessType', 'total_license_application']
```

```python
df_x
```

```python
df_1 = df.groupby(["FOLDERYEAR","BusinessIndustry",'BusinessType'],as_index=False)[["label"]].sum()
df_1.columns = ['FOLDERYEAR', 'BusinessIndustry','BusinessType', 'total_license_issued']
```

```python
#df_1.head(30)
```

```python
#Remove 2010 Winter games
#df_1 = df_1[df_1.BusinessIndustry != 'Historic']
df_1 = df_1[df_1.FOLDERYEAR != 1996]
df = df[df.FOLDERYEAR != 1996]
```

```python
#df_1.head(30)
```

```python
industry_grouped = df.groupby('BusinessIndustry',as_index=False )['label'].sum()
industry_grouped.columns = ['BusinessIndustry','total_license_issued']
```

```python
industry_grouped
```

```python
year_grouped_1 = df_x.groupby('FOLDERYEAR',as_index=False )['total_license_application'].sum()
year_grouped_1.columns = ['Year','total_license_application']
```

```python
merged = pd.merge(year_grouped, year_grouped_1, on='Year', how='outer')
```

```python

merged['license_not_issued'] = merged.apply(lambda row: ((row.total_license_application) - (row.total_license_issued)), axis = 1)
```

```python
merged
```

```python
bars_n = alt.Chart(merged).mark_bar(opacity=0.5).encode(
     x=alt.X('Year',type = "ordinal", title=None, axis=alt.Axis(grid=False)),
     y=alt.Y("license_not_issued:Q",title=None)
   )



text_n = bars_n.mark_text(
    align='left',
    baseline='top',
    dy=-20,
    dx=5
).encode(
    text='license_not_issued:Q'
)



line_n = bars_n.mark_line(color='red').encode(
    y="license_not_issued:Q"
)

#(bar + line).properties(width=600)


#bars_1
plot_x = (bars_n +line_n).properties(title= 'Total Unissued Licenses',width=700, height=500).configure_axis(labels=False, labelFontWeight = 100, labelFontSize = 10).configure_axis(
    grid=False
).configure_view(
    strokeWidth=0
).configure_title(
    fontSize=20,
    font='Arial',
    anchor='middle',
    color='black'
)
plot_x
```
