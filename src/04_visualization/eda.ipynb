```python
df_1 = df.groupby(["FOLDERYEAR","BusinessIndustry",'BusinessType'],as_index=False)[["label"]].sum()
df_1.columns = ['FOLDERYEAR', 'BusinessIndustry','BusinessType', 'total_license_issued']
```

```python
#df_1.head(30)
```

```python
#Remove 2010 Winter games
df_1 = df_1[df_1.BusinessIndustry != 'Historic']
```

```python
#df_1.head(30)
```

```python
industry_grouped = df.groupby('BusinessIndustry',as_index=False )['label'].sum()
industry_grouped.columns = ['BusinessIndustry','total_license_issued']
```

```python
year_grouped = df.groupby('FOLDERYEAR',as_index=False )['label'].sum()
year_grouped.columns = ['Year','total_license_issued']
```

```python
#year_grouped
```

```python
bars_1 = alt.Chart(year_grouped, title='Total License Issued').mark_bar(opacity=0.6).encode(
     x=alt.X("total_license_issued:Q",title = "Licenses issued",axis=None),
     y=alt.Y('Year',type = "ordinal")
   )

text_1 = bars_1.mark_text(
    align='left',
    baseline='middle',
    dx=8  # Nudges text to right so it doesn't appear on top of the bar
).encode(
    text='total_license_issued:Q'
)

(bars_1 + text_1).properties(width=700, height=500).configure_axis(titleFontSize=18, titleFontWeight = 500, labelFontWeight = 100, labelFontSize = 12
        )
```

```python
## Remove 2010 wintergames history
industry_grouped = industry_grouped[industry_grouped.BusinessIndustry != 'Historic']
```

```python
bars_2 = alt.Chart(industry_grouped, title='Total License Issued').mark_bar(opacity=0.7).encode(
     x=alt.X("total_license_issued:Q", title = "Licenses issued",axis=None),
     y=alt.Y('BusinessIndustry:N',title = "Industry", sort='-x')
   )

bars_2

text_2 = bars_2.mark_text(
    align='left',
    baseline='middle',
    dx=8  # Nudges text to right so it doesn't appear on top of the bar
).encode(
    text='total_license_issued:Q'
)

(bars_2 + text_2).properties(width=700, height=500).configure_axis(titleFontSize=18, titleFontWeight = 500, labelFontWeight = 100, labelFontSize = 12
        )
```
